```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/instagram_data.csv")
df
```

```python
df['filename'] = df['image_path'].apply(lambda x: x.replace("../Data/", "./data/"))
df
```

```python
df['t_zeroed'] = (df['t'] - df['t'].min())//1000
print(df['t'].max() - df['t'].min())
```

```python
# shuffle data
df = df.sample(frac=1).reset_index(drop=True)
```

```python
# standardize data
df['t_zeroed'] = (df['t_zeroed'] - df['t_zeroed'].mean()) / df['t_zeroed'].std()
df['no_of_comments'] = (df['no_of_comments'] - df['no_of_comments'].mean()) / df['no_of_comments'].std()
df['follower_count_at_t'] = (df['follower_count_at_t'] - df['follower_count_at_t'].mean()) / df['follower_count_at_t'].std()
#save std and mean of likes to a file
mean_likes = df['likes'].mean()
std_likes = df['likes'].std()
np.save('mean_likes.npy', mean_likes)
np.save('std_likes.npy', std_likes)
df['likes'] = (df['likes'] - df['likes'].mean()) / df['likes'].std()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
```

```python
# split train into train and validation
train_df, val_df = train_test_split(train_df, test_size=0.2, random_state=42)
```

```python
train_df.to_csv("train.csv", index=False)
val_df.to_csv("val.csv", index=False)
test_df.to_csv("test.csv", index=False)
```
